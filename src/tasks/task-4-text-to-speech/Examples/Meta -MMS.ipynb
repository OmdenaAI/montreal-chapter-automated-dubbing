```python
import urllib.request
import tarfile

def download(lang):
    url = f"https://dl.fbaipublicfiles.com/mms/tts/{lang}.tar.gz"
    file_name = f"{lang}.tar.gz"

    urllib.request.urlretrieve(url, file_name)

    with tarfile.open(file_name, "r:gz") as tar:
        tar.extractall("data")

```

```python
# Usage example
download("tam")
```

```python
download("spa")
```

```python
download("fra")
```

```python
download("deu")
```

```python
Audio(wav["x"],rate=wav["sampling_rate"])
```
