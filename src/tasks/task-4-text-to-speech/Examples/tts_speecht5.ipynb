Import necessary modules and initialize text-to-speech model

```python
from transformers import SpeechT5Processor, SpeechT5ForTextToSpeech
processor = SpeechT5Processor.from_pretrained("microsoft/speecht5_tts")
model = SpeechT5ForTextToSpeech.from_pretrained("microsoft/speecht5_tts")
```

 Prepare input text for synthesis

```python
inputs = processor(text="Saving your Search history helps you rediscover relevant results, pick up where you left off, and get a more personalized experience in Search and across other Google services, like Maps and YouTube.", return_tensors="pt")
```

Load speaker embeddings dataset and extract speaker embeddings

```python
from datasets import load_dataset
embeddings_dataset = load_dataset("Matthijs/cmu-arctic-xvectors", split="validation")

import torch
speaker_embeddings = torch.tensor(embeddings_dataset[7306]["xvector"]).unsqueeze(0)
```

Initialize vocoder model and generate speech

```python
from transformers import SpeechT5HifiGan
vocoder = SpeechT5HifiGan.from_pretrained("microsoft/speecht5_hifigan")
```

```python
speech = model.generate_speech(inputs["input_ids"], speaker_embeddings, vocoder=vocoder)
```

```python
from IPython.display import Audio

Audio(speech, rate=16000)
```
