# Import Libraries

```python
import scrapetube
from pytube import YouTube 
import moviepy.editor
import pandas as pd
```

# Extracting Video Urls from Channel

```python
# Channel Name
channel_url = "https://www.youtube.com/@TEDEdEspanol"

# Number of Videos
nr_videos = 250

# Youtube Scraping 
videos = scrapetube.get_channel(channel_url = channel_url, limit = nr_videos)

# Video Url List
url_list = []

# Extract all Urls
for video in videos:
    video_link="https://www.youtube.com/watch?v="+str(video['videoId'])
    url_list.append(video_link)

    
```

# Extracting Length and Title of each Video

```python
# Duration List in Seconds
length_list = []

# Video Title List
vid_title_list = []

for i in url_list: 
  video = f"{i}" 

  yt = YouTube(video)         ## this creates a YOUTUBE OBJECT
  video_length = yt.length    ## this extract the video length
  vid_title = yt.title        ## this extract the video title
  

  length_list.append(video_length)
  vid_title_list.append(vid_title)
```

```python
# Check Number of Videos
len(length_list)
```

# Creating DataFrame

```python
df = pd.DataFrame({'Url':url_list,'Length_Seconds':length_list, "Video_Title": vid_title_list})
df = df.sort_values("Length_Seconds", ascending=True)
df
```

```python
# Write DataFrame to Excel file
df.to_excel('/content/drive/MyDrive/Colab Notebooks/01 Omdena/Canada - Dubbing/Task 1 Preprocessing/TEDEdEspanol.xlsx')
```

# Extracting Audio and Video

```python
# Download the first 2 Youtube Videos and Audios of the dataframe (change it if required)

for i in df.Url[:2]:

  video_extension = "mp4"
  audio_extension = "mp3"

  yt = YouTube(f"{i}")

  yt.streams.filter(progressive=True, file_extension = video_extension) \
  .order_by('resolution').desc().first().download(filename = f"{yt.title}.{video_extension}")

  # Load the Video (change path name if required)
  video = moviepy.editor.VideoFileClip(filename = f"/content/{yt.title}.{video_extension}")

  # Extract Audio if Audio is Available
  if video.reader.infos["audio_found"] == True:

    #Extract the Audio
    audio = video.audio

    #Export the Audio to mp3 format
    audio.write_audiofile(filename = f"/content/{yt.title}.{audio_extension}")

  else:
    continue
  
```
