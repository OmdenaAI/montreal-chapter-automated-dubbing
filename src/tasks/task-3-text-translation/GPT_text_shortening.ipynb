````python
import os
import openai
````

````python
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv('drive/MyDrive/Omdena_text_translation/.env')) # read local .env file
openai.api_key = os.getenv('OPENAI_API_KEY')
````

````python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0,
    )
    return response.choices[0].message["content"]
````

````python
text = f"""
Koalas typically inhabit open Eucalyptus woodland, as the leaves of these trees make up most of their diet.
"""
prompt = f"""
Shorten the text delimited by triple backticks by 20%.
```{text}```
"""
response = get_completion(prompt)
print(response)
````

````python
text2 = f"""
Once upon a time there was an old mother pig who had three little pigs and not enough food to feed them. So when they were old enough, she sent them out into the world to seek their fortunes.
"""
prompt2 = f"""
Make the text delimited by triple backticks shorter.
```{text2}```
"""
response2 = get_completion(prompt2)
print(response2)
````
