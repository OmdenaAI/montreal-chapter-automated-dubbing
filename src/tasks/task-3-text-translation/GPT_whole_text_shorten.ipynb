```python
import os
import openai
```

```python
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv('drive/MyDrive/Omdena_text_translation/.env')) # read local .env file
openai.api_key = os.getenv('OPENAI_API_KEY')
```

```python
text = f"""
duration: 02.200, text_segment:  Most of my free time in these 10 years
duration: 04.160, text_segment:  has been dedicated to make plastic bricks move.
duration: 03.440, text_segment:  And this is all time that I will never get back.
duration: 05.760, text_segment:  So I started to ask myself, is all of this wasted time?
duration: 02.080, text_segment:  We're always being told to maximize our potential,
duration: 02.140, text_segment:  to live life to the fullest.
duration: 04.240, text_segment:  And oftentimes, that seems to include not wasting our time.
duration: 01.560, text_segment: I mean, we like scrolling through social media.
duration: 03.240, text_segment: We like watching countless hours of television shows.
duration: 03.560, text_segment:  But sometimes, we might blame ourselves for that afterwards.
"""

delimeter = "####"

system_message = f"""You are tasked with dubbing a video. You will be provided with a translated text delimited by {delimeter}
that has a form of a list, where each item consists of time duration and text segment to be said during this time interval. Follow these
steps to shorten the text segments to fit the provided time intervals.

Step 1:{delimeter} Read all the text segments to gain the understanding of the whole text.

Step 2:{delimeter} Provide a condensed version of each text segment that accurately conveys the same meaning. Make sure that
the number of tokens used in the text segment you return does not exceed the original number of tokens used.

Step 3:{delimeter} With the exception of very short segments, make sure that the number of tokens of the condensed text segment
is proportional to the time duration.

Step 4:{delimeter} Return your answer in the same format as the original text, i.e a list of items containing time duration and
text segment. Make sure that you return the same number of items in the list as given. Retain the duration, only change text
segment. """

user_message = f"""{delimeter}{text}{delimeter}"""
```

```python
messages = messages =  [
{'role':'system',
 'content': system_message},
{'role':'user',
 'content': user_message},
]
```

```python
def get_completion_from_messages(messages,
                                 model="gpt-3.5-turbo",
                                 temperature=0):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature,
    )
    return response.choices[0].message["content"]
```

```python
response = get_completion_from_messages(messages)
print(response)
```
