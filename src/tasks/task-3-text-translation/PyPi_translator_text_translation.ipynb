```python
from translate import Translator
translator = Translator(to_lang="fr")
```

```python
text = 'One, two, three, four, five, once I caught a fish alive, six, seven, eight, nine, ten, then I let go again.'
```

```python
translation = translator.translate(text)
translation
```

```python
translator2 = Translator(to_lang='en', from_lang='fr')
```

```python
translation2 = translator2.translate(translation)
translation2
```

```python
text2 = 'Hiring accelerated for the second straight month in May, in the face of higher interest rates and elevated inflation.'
```
